```python
import pandas as pd

url = "https://github.com/gastonstat/CreditScoring/raw/master/CreditScoring.csv"
#!wget $url -O CreditScoring.csv
df = pd.read_csv('CreditScoring.csv')
df.head()
```

```python
df.columns= df.columns.str.lower()
df.head()
```

EDA

```python
df.status.value_counts()
```

```python
df.status =df.status.map({1:'ok', 2:'default',0:'unk'})
```

```python
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}

df.home = df.home.map(home_values)

marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}

df.marital = df.marital.map(marital_values)

records_values = {
    1: 'no',
    2: 'yes',
    0: 'unk'
}

df.records = df.records.map(records_values)

job_values = {
    1: 'fixed',
    2: 'partime',
    3: 'freelance',
    4: 'others',
    0: 'unk'
}

df.job = df.job.map(job_values)
```

```python
df.replace(99999999.0, np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
#remove rows with missing values
df=df[df.status != np.nan].reset_index(drop=True)
```

```python
df.describe().round()
```

#spliting the data into train test validation 

```python
from sklearn.model_selection import train_test_split

# Split the data into train (60%), validation (20%), and test (20%) sets
train_df, temp_df = train_test_split(df, test_size=0.4, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# Display the sizes of the splits
print(f"Train set size: {train_df.shape[0]}")
print(f"Validation set size: {val_df.shape[0]}")
print(f"Test set size: {test_df.shape[0]}")
```

```python
y_train = (train_df.status == 'default').astype(int).values
y_val = (val_df.status == 'default').astype(int).values
y_test = (test_df.status == 'default').astype(int).values
train_df = train_df.drop(columns=['status'])
val_df = val_df.drop(columns=['status'])
test_df = test_df.drop(columns=['status'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import classification_report, accuracy_score
```

```python
#Prepare the feature columns
feature_cols = ['seniority', 'time', 'age', 'expenses', 'income', 'assets', 'debt', 'amount', 'price']
# Create the vectorizer
dv = DictVectorizer(sparse=False)
dv.fit(train_df[feature_cols].to_dict(orient='records'))
```

```python
# Prepare the training and validation data
X_train = dv.transform(train_df[feature_cols].to_dict(orient='records'))
X_val = dv.transform(val_df[feature_cols].to_dict(orient='records'))
X_test = dv.transform(test_df[feature_cols].to_dict(orient='records'))
```

```python
# Initialize the DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42, max_depth=200,max_leaf_nodes=10,max_features=5)
# Train the model
clf.fit(X_train, y_train)
# Predict on the validation set
y_val_pred = clf.predict(X_val)


# Validate the result
print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("Classification Report:\n", classification_report(y_val, y_val_pred))
# Predict on the test set
y_test_pred = clf.predict(X_test)


# Test the result
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=42, n_estimators=1000, max_depth=25, max_features='sqrt')

# Train the model
rf_clf.fit(X_train, y_train)

# Predict on the validation set
y_val_pred_rf = rf_clf.predict(X_val)

# Validate the result
print("Validation Accuracy (Random Forest):", accuracy_score(y_val, y_val_pred_rf))
print("Classification Report (Random Forest):\n", classification_report(y_val, y_val_pred_rf))

# Predict on the test set
y_test_pred_rf = rf_clf.predict(X_test)

# Test the result
print("Test Accuracy (Random Forest):", accuracy_score(y_test, y_test_pred_rf))
print("Classification Report (Random Forest):\n", classification_report(y_test, y_test_pred_rf))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [10, 20, 100, 200],
    'max_depth': [10, 25],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize the RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf_clf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best parameters found: ", best_params)

# Train the model with the best parameters
best_rf_clf = RandomForestClassifier(**best_params, random_state=42)
best_rf_clf.fit(X_train, y_train)

# Predict on the validation set
y_val_pred_best_rf = best_rf_clf.predict(X_val)

# Validate the result
print("Validation Accuracy (Tuned Random Forest):", accuracy_score(y_val, y_val_pred_best_rf))
print("Classification Report (Tuned Random Forest):\n", classification_report(y_val, y_val_pred_best_rf))

# Predict on the test set
y_test_pred_best_rf = best_rf_clf.predict(X_test)

# Test the result
print("Test Accuracy (Tuned Random Forest):", accuracy_score(y_test, y_test_pred_best_rf))
print("Classification Report (Tuned Random Forest):\n", classification_report(y_test, y_test_pred_best_rf))
```

```python
# Extract the results from the GridSearchCV object
results = pd.DataFrame(grid_search.cv_results_)

# Pivot the results to create a matrix for the heatmap
pivot_table = results.pivot_table(values='mean_test_score', 
                                  index='param_max_depth', 
                                  columns='param_n_estimators')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='viridis')
plt.title('Grid Search CV Results')
plt.xlabel('Number of Estimators')
plt.ylabel('Max Depth')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.impute import SimpleImputer

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_val_imputed = imputer.transform(X_val)
X_test_imputed = imputer.transform(X_test)

# Initialize the GradientBoostingClassifier
gb_clf = GradientBoostingClassifier(random_state=42, n_estimators=500, learning_rate=0.05, max_depth=10, min_samples_leaf=5,max_features='sqrt')

# Train the model
gb_clf.fit(X_train_imputed, y_train)

# Predict on the validation set
y_val_pred_gb = gb_clf.predict(X_val_imputed)

# Validate the result
print("Validation Accuracy (Gradient Boosting):", accuracy_score(y_val, y_val_pred_gb))
print("Classification Report (Gradient Boosting):\n", classification_report(y_val, y_val_pred_gb))

# Predict on the test set
y_test_pred_gb = gb_clf.predict(X_test_imputed)

# Test the result
print("Test Accuracy (Gradient Boosting):", accuracy_score(y_test, y_test_pred_gb))
print("Classification Report (Gradient Boosting):\n", classification_report(y_test, y_test_pred_gb))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [10,60,120],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [1,2,3,5],
    'min_samples_leaf': [1, 5],
}

# Initialize the GradientBoostingClassifier
gb_clf = GradientBoostingClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=gb_clf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train_imputed, y_train)

print (grid_search.best_params_)

```

```python
best_params = grid_search.best_params_
gb_tuned = GradientBoostingClassifier(**best_params)
gb_tuned.fit(X_train_imputed, y_train)
y_val_pred_best_rf= gb_tuned.predict(X_val_imputed)
y_test_pred_best_rf = gb_tuned.predict(X_test_imputed)

# Validate the result
print("Validation Accuracy (Tuned Gradient Boosting):", accuracy_score(y_val, y_val_pred_best_rf))
print("Classification Report (Tuned Gradient Boosting):\n", classification_report(y_val, y_val_pred_best_rf))   
#test the result
print("Test Accuracy (Tuned Gradient Boosting):", accuracy_score(y_test, y_test_pred_best_rf))

print("Classification Report (Tuned Gradient Boosting):\n", classification_report(y_test, y_test_pred_best_rf))
```
