```python
# we always mix the data before splitting it 
np.random.seed(2)
idx=np.arange(len(df))
np.random.shuffle(idx)
idx
```

```python
n=len(df)
n_val=int(0.2*n)
n_test=int (0.2*n)
n_train= n-n_val-n_test
df_train=df.iloc[idx[:n_train]]
df_val=df.iloc[idx[n_train:n_train+n_val]]
df_test=df.iloc[idx[n_val+n_train:]]
df_train.head()


```

```python
df_train=df_train.reset_index(drop=True)
df_val=df_val.reset_index(drop=True)
df_test=df_test.reset_index(drop=True)
df_train.head()
```

Creating Y variabales and deleting it from the tables 

```python
y_train= df_train.msrp.values
y_test=df_test.msrp.values
y_val=df_val.msrp.values
del df_train['msrp']
del df_test['msrp']
del df_val['msrp']

```

```python
df_train.shape[0]
```

Training Linear regression model

```python
def train_LR (X,y):
    X =np.column_stack([np.ones(X.shape[0]),X])
    XTI= np.linalg.inv(X.transpose().dot(X)+np.eye(X.shape[1])*0.0001)
    w_full= XTI.dot(X.T).dot(y)
    return (w_full[0], w_full[1:])

def prepare_X(df,base):
    df_num = df[base]
    df_num = df_num.fillna(0)
    X = df_num.values
   
    return X
```

```python
base = ['year','engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg', 'popularity']

df_train['nb_doors_2']=(df_train.number_of_doors==2).astype('int')
df_train['nb_doors_3']=(df_train.number_of_doors==3).astype('int')
df_train['nb_doors_4']=(df_train.number_of_doors==4).astype('int')
base.__add__(['nb_doors_2','nb_doors_3','nb_doors_4'])

df_x=df_train[base]
X=  df_x.fillna(0).values 

print (X[0:4])

w0, w= train_LR(X,np.array(y_train))

```

```python
print(w0, w)
```

```python
y_predict=w0+X.dot(w) 
print ( np.expm1(y_train[10]))
print ('predicted price:', np.expm1(y_predict[10]) )
```

```python
sns.histplot((y_predict,y_train),bins=50)
```

RMSE

```python
def rmse_calcul(y_tr,y_pr):
    # Calculate the mean squared error (MSE) by taking the mean of the squared differences
    meanSquaredError = ((y_pr - y_tr) ** 2).mean()
    # Calculate the RMSE by taking the square root of the MSE
    rmse = np.sqrt(meanSquaredError)
    return rmse
```

```python
print (rmse_calcul(y_train,y_predict))
```

Validation of the model

```python
X_val=prepare_X(df_val,base)
y_pr_val=w0+X_val.dot(w)
print ('error estimate:',rmse_calcul(y_val,y_pr_val)) 
```

2.12 - Categorical Variables

```python
new_df_train=df_train.copy()
categorical_var=df_train.dtypes[df_train.dtypes=='object'].index
added_features=[]
for category in categorical_var:
    #get the most used 5 values in each category
    best_5=df_train[category].value_counts().head(10).index
    for best in best_5:
        #added_features.append(
        new_category=category+'_'+best
        added_features.append(new_category)
        print (new_category)
        new_df_train[new_category]=(df_train[category]==best).astype('int')

    #
```

```python
new_based=base.__add__(added_features)
X_train= prepare_X(new_df_train,new_based)
w0,w=train_LR(X_train,y_train)
y_predict=w0+X_train.dot(w)
print(rmse_calcul(y_train,y_predict))
sns.histplot((y_predict,y_train),bins=50)
```
